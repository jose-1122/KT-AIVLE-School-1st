```python
### 모듈, 함수 불러오기
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import font_manager, rc                          # 한글이 깨지지 않도록 설정
font_loc = "C:/Windows/Fonts/malgunbd.ttf"                        # 맑은 고딕
font_name = font_manager.FontProperties(fname=font_loc).get_name()
matplotlib.rc('font', family=font_name)

### 그래프
fig = plt.figure(figsize=(10,10))                                 # 크기 설정
ax = fig.add_subplot(projection='3d')                             # 3차원
plt.xlim(-1225, 1225)                                             # x축 최솟값, 최댓값
plt.ylim(-1225, 1225)                                             # y축 최솟값, 최댓값
ax.grid(False)                                                    # 격자 off
ax.set_title('[ K T ]   A I V L E   S c h o o l', c='red', fontsize=25, fontweight='bold') # 제목
ax.set_xlabel('X 축 ; [-1225, 1225]', labelpad=10, fontsize='13') # x축 레이블
ax.set_ylabel('Y 축 ; [-1225, 1225]', labelpad=10, fontsize='13') # y축 레이블
ax.set_zlabel('z 축 ; [0, 800]',      labelpad=10, fontsize='13') # z축 레이블
ax.text(-2400,0,740, 'Wish', c='green', fontsize='40', fontweight='bold') # 그래프 위 텍스트 1
ax.text(-1350,0,780, "y'all", c='red', fontsize='40', fontweight='bold')  # 그래프 위 텍스트 2
ax.text(-550,0,910, 'Merry', c='green', fontsize='40', fontweight='bold') # 그래프 위 텍스트 3
ax.text(550,0,950, 'X-mas', c='red', fontsize='40', fontweight='bold')    # 그래프 위 텍스트 4
ax.text(1600,0,1110, '(:', c='green', fontsize='40', fontweight='bold')   # 그래프 위 텍스트 5

###크리스마스 트리
dict_x = {111:'green', 148:'lightgreen', 185:'yellowgreen',
          222:'green', 259:'lightgreen', 296:'yellowgreen',
          333:'green', 370:'lightgreen', 407:'yellowgreen',
          444:'green', 481:'lightgreen', 518:'yellowgreen',
          555:'green', 592:'lightgreen', 629:'gold',
          666:'green', 703:'lightgreen', 740:'brown',
          777:'green'}
for k in dict_x.keys():
    X = [np.cos(i/2)*(k-i) for i in range(k)]
    Y = [np.sin(i/9)*(k-i) for i in range(k)]
    Z = [i for i in range(k)]
    ax.scatter(X,Y,Z, s=50, c=dict_x[k], marker='.')

### 크리스마스 트리 장식
ax.scatter(-10,0,k+33, s=k, c='deeppink', marker='*')             # 1. 별

step=7                                                            # 2. 장식a
X = [np.cos(i/7)*(k-i) for i in range(1, k, step)]
Y = [np.sin(i/7)*(k-i) for i in range(1, k, step)]
Z = [i for i in range(1, k, step)]
ax.scatter(X,Y,Z, s=60, c='yellow', marker='o')

step=14                                                           # 3. 장식b
X = [np.cos(i/7)*(k-i) for i in range(1, k, step)]
Y = [np.sin(i/7)*(k-i) for i in range(1, k, step)]
Z = [i for i in range(1, k, step)]
ax.scatter(X,Y,Z, s=40, c='cyan', marker='^')

step=7                                                            # 4. 장식c
X = [np.sin(i/7)*(k-i) for i in range(1, k, step)]
Y = [np.cos(i/7)*(k-i) for i in range(1, k, step)]
Z = [i for i in range(1, k, step)]
ax.scatter(X,Y,Z, s=60, c='deeppink', marker='p')

### 눈
color = ['azure', 'skyblue', 'aquamarine', 'lavender', 'darkseagreen']
for c in color:
    X = np.random.randint(-1225, 1225, size=40)
    Y = np.random.randint(-1225, 1225, size=40)
    Z = np.random.randint(0, 800, size=40)
    ax.scatter(X, Y, Z, c=c, marker='x')

plt.show()                                                        # 화면에 출력
```
